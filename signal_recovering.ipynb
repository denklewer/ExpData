# Восстановление сильно зашумленного сигнала

```python
import grapher
from analysis import FT, examples, trends, preprocessing, statistics
import prng

r = prng.PRNG()
%matplotlib inline
```

Установим частоту дискретизации и количество семплов

```python
deltaT = 0.001
N = 1000
```

Также установим свойства синусодиной функции и константу, насколько шум превышает сигнал

```python
f0 = 7
A0 = 100

D = 10
f = examples.sin(A0=A0, f0=f0, dt=deltaT)
```

```python
# Многократно повторяем эксперимент и получаем М-канальную систему, затем все складываем и усредняем
M = 1000
values = [0] * N

for i in range(M):
	values = [values[t] + f(t) + r.uniform(-1, 1) * D * A0 for t in range(N)]

values = [values[t] / M for t in range(N)]
```

```python
grapher.init(1)
grapher.set_subplot(1, x_arr=[i * deltaT for i in range(N)],
                    y_arr=values, title="function", subplot_y_number=1, subplot_x_number=1)
grapher.show()
```
